<hr><div class="alert alert-block alert-danger" style="margin-top: 20px">
<h3>loc[ ] & iloc[ ]</h3></div><hr>

* loc[]: Used for label-based indexing to access a group of rows and columns<br>
by labels or a boolean array.
* iloc[]: Used for integer-based (position-based) indexing to access rows and columns<br>
by their integer positions.

```python
data.loc[2]  
```

```python
data.loc[[0,3,7,9]]  # must double [[]]
```

```python
#df.iloc[start_row : end _row , start_col: end_col ] we can select particular rorws and colums

subset = data.loc[ : , ['year', 'pop']]  # [only year,pop column]

#subset = data.loc[ : , 'year': 'pop']  # year to pop
```

```python
subset.head()
```

```python
data.iloc[-1]  #data.loc[-1] doesn't work
```

```python
data.iloc[[0,-1,3,-7,9]] # must double [[]]
```

```python
#df.iloc[start_row : end _row , start_col: end_col ]  we can select particular rorws and colums
subset = data.iloc[:, [2, 4]]  #2 & 4 only
#subset = data.iloc[:, 2: 4] #  2 to 4 
```
