```python
weather = pd.read_csv("pandas_data/weather.csv") # another data
```

```python
weather
```

```python
weather_long = weather.melt(id_vars  = ['id', 'year', 'month', 'element'] , var_name = 'day', value_name = 'temp')
```

```python
weather_long.head()
```

```python
(
    weather_long
        .pivot_table(index=['id', 'year', 'month', 'day'],columns='element',values='temp')
        # .reset_index()
)
```

```python
(
    weather_long
        .pivot_table(index=['id', 'year', 'month', 'day'],columns='element',values='temp')
        .reset_index()
)
```
