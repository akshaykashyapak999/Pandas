<hr><div class="alert alert-block alert-danger" style="margin-top: 20px"><h3>Facetgrid</h3>
    <code>Facetgrid</code>
    in Seaborn allows you to create a <code>grid of plots</code> based on subsets<br>
    of your dataset, using categorical variables for arranging multiple plots in<br>
    rows and columns. It's useful for visualizing relationships across multiple<br>
    dimensions of your data simultaneously.
</div><hr>

```python
facet = sns.FacetGrid(dataset, col='time', row='smoker', hue='sex')
#empty container or defining how many plot will be there (using row & col)
```

```python
# comment/uncomment (see what happen)
facet = sns.FacetGrid(dataset, col='time', row='smoker', hue='sex') 
# why it not coming from perivious code,
# because its already rendered (in editor neednot to write again)

facet.map(plt.scatter, 'total_bill', 'tip') 
```
